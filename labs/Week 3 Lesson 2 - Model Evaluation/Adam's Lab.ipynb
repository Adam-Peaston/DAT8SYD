## Here's how to find a string from inside another string inside a pandas dataframe

```python
dummy = pd.Series(['fghjkjhgfghjkjhgkillmekjhgfghjkljhgf','FGHJKJHGFDSFGHJ'], index=[0,1])
dummy.str.find('killme', start=0, end=None)
```
