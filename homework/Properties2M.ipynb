```python
#import pandas and drag in the data
import pandas as pd
prop_pred = pd.read_csv('properties_2016.csv')
prop_targ = pd.read_csv('train_2016.csv')
```

```python
#merge the two datasets
prop_data = prop_pred.merge(prop_targ, how='outer', on='parcelid')
prop_data.shape
```

```python
#set the index as parcelid
prop_data.set_index('parcelid', inplace=True)
prop_data.head()
```

```python
#some basic facts about the data
prop_data.describe()
```

```python
#trim dataset to only those entries for which logerror is available ~90,000 entries
prop_data_letrim = prop_data.dropna(axis=0, how='any', subset=['logerror'])
prop_data_letrim.shape
```

```python
#count how many entries in each feature survived
entries = prop_data_letrim.count()
entries.sort_values(inplace=True)
```

```python
#show me what that looks like
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
 
objects = entries.index
y_pos = np.arange(len(objects))
outcomes = entries
 
plt.figure(figsize=(15,15))
plt.barh(y_pos, outcomes, 0.8, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('Entry Count')
plt.title('Entries available')
 
plt.show()
```

```python
#pull transaction month out of datestamp
prop_data_letrim['transactionmonth'] = prop_data_letrim['transactiondate'].str[5:7]
prop_data_letrim[['transactiondate','transactionmonth']]
```
